### Praktikum 1 tipe data dasar

```python
# 1. Number
# integer 
angka_int = 10
print("angka_int =", angka_int, "-->", type(angka_int))

# float 
angka_float = 3.14
print("angka_float = ", angka_float, "-->", type(angka_float))

# compex 
angka_compex = 2 + 3j
print("angka_compex = ", angka_compex, "-->", type(angka_compex))


# 2. Boolean 
is_active = True
print("is_active = ", is_active, "-->", type(is_active))

# 3. String 
teks = "Hello, Python!!!"
print("teks = ", teks, "-->", type(teks))

# 4. List
daftar_angka = [1, 2, 3, 4, 5]
print("daftar_angka", daftar_angka, "-->", type(daftar_angka))

# 5. Tuple
koordinat = (10,20)
print("koordinat", koordinat, "-->", type(koordinat))

# 6. Dictionary
data_mahasiswa = {
    "nama" : "Andi",
    "nim" : "4.33.24.2.18",
    "jurusan" : "Teknik Informatika"
}
print("data_mahasiswa = ", data_mahasiswa, "-->", type(data_mahasiswa))

# 7. set
himpunan_angka = {1, 2, 3, 2, 1}
print("himpunan_angka = ", himpunan_angka, "-->", type(himpunan_angka))

# 8. Contoh penggunaan konversi tipe data
nilai_str = "100"
print("\nnilai_str = ", nilai_str, "-->", type(nilai_str))
nilai_int = int(nilai_str)
print("\nnilai_int = ", nilai_int, "-->", type(nilai_int))

```

### Praktikum 2 Variable dan Operasi

```python
# 1. Pendeklarasian variable
nama = "Budi"
umur = 20
tinggi = 170.1
is_student = True

print("Nama = ", nama)
print("Umur =", umur)
print("Tinggi = ", tinggi, "cm")
print("Mahasiswa = ", is_student)

# 2. Operasi Aritmatika
a = 10
b = 3

penjumlahan = a + b
pengurangan = a - b
perkalian = a * b
pembagian = a / b
pembagian_bulat = a // b
pangkat = a ** b
modulus = a % b

print("\nOPERASI ARITMATIKA")
print("a = ", a, " b = ", b)
print("penjumlahan = ", penjumlahan)
print("pengurangan = ", pengurangan)
print("perkalian = ", perkalian)
print("pembagian = ", pembagian)
print("pembagian_bulat = ", pembagian_bulat)
print("modulus = ", modulus)
print("pangkat = ", pangkat)

# 3. Operasi perbandingan
lebih_besar = a > b
kurang_dari = a < b
sama_dengan = a == b
tidak_sama = a != b
lebih_besar_sama = a >= b
kurang_sama = a <= b

print("\nOPERASI PERBANDINGAN")
print("a > b = ", lebih_besar)
print("a < b = ", kurang_dari)
print("a == b = ", sama_dengan)
print("a != b = ", tidak_sama)
print("a >= b = ", lebih_besar_sama)
print("a <= b = ", kurang_sama)

# 4. Operasi logika 
x = True
y = False

logika_and = x and y
logika_or = x or y
logika_not_x = not x

print("\nOPERASI LOGIKA")
print("x = ", x, " y = ", y)
print("x and y = ", logika_and)
print("x or y = ", logika_or)
print("not x = ", logika_not_x)

# 5. contoh di dalam percabangan
if a > b and b > 0 :
    print("\nKondisi terpenuhi")
else :
    print("Kondisi tidak terpenuhi")
```

### Praktikum 3. Percabangan 

```python
# 1. If sederhana

nilai = 85
if nilai > 80:
    print("Anda Lulus dengan nilai tinggi\n")


# 2. IF ELSE

umur = 17
if umur >= 18 :
    print("Anda cukup umur untuk membuat SIM")
else :
    print("Anda belum cukup umur unutk membuat SIM\n")


# 3. IF ELSE ELIF

hari = "rabu"
if hari == "senin":
    print("Hari senin - saatnya kembali bekerja!")
elif hari == "selasa":
    print("Hari selasa - Jadwal rapat mingguan")
elif hari == "rabu":
    print("Hari rabu - Ada diskon di beberapa toko")
else :
    print("Hari lainnya - atur jadwalmu dengan baik\n")

# 4. IF bersarang

suhu = 35
if suhu > 30:
    print("Cuaca cukup panas")
    if suhu > 40:
        print("Bahkan sangat terik! Disarankan banyak minum air")
    else : 
        print("Masih relatif normal, tapi tetap jaga kesehatan juga\n")
else :
    print("Cuaca sepertinya cukup sejuk\n")

# 5. Menggabungkan Percabangan dengan Operasi Logika 

nilai_teori = 75 
nilai_praktik = 80
if nilai_teori >= 70 and nilai_praktik > 70 :
    print("Anda Lulus karena nilai teori dan praktik memadai\n")
elif nilai_teori > 70 and nilai_praktik > 70 :
    print("Anda perlu meningkatkan nilai teori dan praktik")
elif nilai_teori > 70 :
    print("Anda perlu meningkatkan nilai teori")
else :
    print("Anda perlu meningkatkan nilai praktik")

# 6. penggunaan if Ternary
angka = -5
status = "Positif" if angka > 0 else "Negatif atau Nol"
print("Angka =", angka, "-->", status)

```

### Praktikum 4. Perulangan 

```python
# 1. FOR Loop dengan range()
for i in range(5):
    print("Perulangan ke-", i)

print()

# 2. FOR Loop untuk mengiterasi List
buah = ["apel", "mangga", "jeruk", "pisang"]
for item in buah :
    print("Buah =", item)

print()

# 3. WHILE Loop
count = 3
while count < 5:
    print("count = ", count)
    count += 1

print()

# 4. BREAK pada perulangan
for i in range(10):
    if i == 3:
        print("LOOP dihentikan pada i = ", i)
        break
    print("i = ", i)

print()

# 5. CONTINUE pada loop 
for i in range(5):
    if i == 2:
        print("Lewati i = ", i, " dengan continue")
        continue
    print("i = ", i)

print()

# 6. NESTED loop
for i in range(3):
    for j in range(2):
        print(f"i = {i} j = {j}")

print()

# 7. memanfaatkan else pada loop 

for x in range(3):
    print("x =", x)
else :
    print("Loop for telah selsai tanpa berak.\n")

y = 0
while y < 3:
    print("y =", y)
    y += 1
else :
    print("Loop while telah selsai tanpa break.\n")

# 8. PASS sebagai pleaceholder
for i in range(3):
    if i == 1 :
        pass
    print("i =", i)
```
