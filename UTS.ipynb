```python
class AlatElektronik():
    def __init__(self,merk, model, status_daya = False):
        self.merk = merk
        self.model = model
        self._status_daya = status_daya

    def nyalakan(self):
        self._status_daya = True
        print(f"Alat berhasil di nyalakan")

    def matikan(self):
        self._status_daya = False
        print(f"Alat di matikan")

    def get_info(self):
        print(f"Alat dengan merk {self.merk} dan model {self.model}")

    
class Televisi(AlatElektronik):
    def __init__(self,merk, model, ukuran_layar):
        super().__init__(merk, model)
        self.ukuran_layar = ukuran_layar

    def get_info(self):
        print(f"Televisi dengan merk {self.merk}, model {self.model} dan ukuran layar {self.ukuran_layar} Inch")
   
    def ganti_chanel(self, chanel):
        if self._status_daya == True :
            print(f"Televisi berhasil di ganti ke chanel {chanel}")
        else:
            print("Televisi sedang dalam keadaaan mati")

class Radio(AlatElektronik):
    def __init__(self, merk, model, frekuensi):
        super().__init__(merk, model)
        self.frekuensi = frekuensi
    
    def get_info(self):
        print(f"Radio dengan merk {self.merk}, model {self.model}, frekuensi {self.frekuensi} MHz")

    def set_frekuensi(self, frekuensi_baru):
        if self._status_daya == True :
            print(f"Radio di set dari frekuensi {self.frekuensi} ke frekuensi {frekuensi_baru} MHz")
        else :
            print("Radio sedang dalam keadaan mati")

if __name__ == "__main__":
    tv = Televisi("Polytron", "90an", 23.8)
    tv.get_info()
    tv.ganti_chanel("Global TV")
    tv.nyalakan()
    tv.ganti_chanel("Global TV")
    tv.matikan()
    print("\n")
    rd = Radio("Panasonic", "80an", 122)
    rd.get_info()
    rd.set_frekuensi(145)
    rd.nyalakan()
    rd.set_frekuensi(145)
    rd.matikan()
```
