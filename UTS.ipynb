```python
class AlatElektronik:
    def __init__(self, merk, model):
        self.merk = merk
        self.model = model
        self._status_daya = False

    def nyalakan(self):
        self._status_daya = True
        print(f"{self.merk} {self.model} dinyalakan.")

    def matikan(self):
        self._status_daya = False
        print(f"{self.merk} {self.model} dimatikan.")

    def get_info(self):
        print(f"Merk: {self.merk}")
        print(f"Model: {self.model}")
        print(f"Status daya: {'Nyala' if self._status_daya else 'Mati'}")


class Televisi(AlatElektronik):
    def __init__(self, merk, model, ukuran_layar):
        super().__init__(merk, model)
        self.ukuran_layar = ukuran_layar

    def ganti_channel(self, channel):
        if self._status_daya:
            print(f"Channel diganti ke {channel}")
            return channel
        else:
            print("Televisi dalam keadaan mati. Tidak bisa ganti channel.")
            return None

    def get_info(self):
        super().get_info()
        print(f"Ukuran layar: {self.ukuran_layar} inch")


class Radio(AlatElektronik):
    def __init__(self, merk, model, frekuensi):
        super().__init__(merk, model)
        self.frekuensi = frekuensi

    def set_frekuensi(self, frekuensi):
        if self._status_daya:
            self.frekuensi = frekuensi
            print(f"Frekuensi diubah ke {frekuensi} MHz")
        else:
            print("Radio dalam keadaan mati. Tidak bisa mengatur frekuensi.")

    def get_info(self):
        super().get_info()
        print(f"Frekuensi: {self.frekuensi} MHz")


# Contoh penggunaan
tv = Televisi("Samsung", "ModelX", 42)
tv.get_info()
tv.ganti_channel(5)  # gagal karena mati
tv.nyalakan()
tv.ganti_channel(5)
tv.get_info()

print("="*30)

radio = Radio("Sony", "ModelY", 101.2)
radio.set_frekuensi(95.6)  # gagal karena mati
radio.nyalakan()
radio.set_frekuensi(95.6)
radio.get_info()
```
